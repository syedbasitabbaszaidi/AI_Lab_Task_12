# Data Pre-Process

```python
encoded_train_df = train_data.copy()
encoded_test_X = test_X.copy()
encoded_test_y = test_y.copy()
```

```python
def fillnaobj(df,col):
    for i in col:
        df[i] = df[i].fillna(df[i].mode()[0])
col1 = ['Cabin','Embarked']
fillnaobj(encoded_train_df,col1)
col2 = ['Cabin']
fillnaobj(encoded_test_X,col2)
```

```python
encoded_train_df.isnull().sum()
```

```python
encoded_test_X.isnull().sum()
```

```python
def fillna_float(df,col):
    for i in col:
        df[i] = df[i].fillna(df[i].mean())
col1 = ['Age']
fillnaobj(encoded_train_df,col1)
col2 = ['Age',"Fare"]
fillnaobj(encoded_test_X,col2)
```

```python
encoded_train_df.isnull().sum()
```

```python
encoded_test_X.isnull().sum()
```

```python
def drop(df,col):
    for i in col:
        df.drop(i,axis = 1,inplace = True)
col = ['PassengerId','Name','Ticket']
drop(encoded_train_df,col)
drop(encoded_test_X,col)
col2 = ['PassengerId']
drop(encoded_test_y,col2)
```

```python
encoded_train_df.info()
```

```python
def to_int64(df,col):
    for i in col:
        df[i] = df[i].astype('int64')
col = ['Age','Fare']
to_int64(encoded_train_df,col)
to_int64(encoded_test_X,col)
```

```python
def encode_col(df,col):
    for i in col:
        data = pd.DataFrame({i:df[i].unique()})
        label = LabelEncoder()
        label.fit(np.ravel(data))
        df[i] = label.transform(df[i])
col = ['Sex','Cabin','Embarked']
encode_col(encoded_train_df,col)
encode_col(encoded_test_X,col)
```

```python
col = ['Sex','Cabin','Embarked']
to_int64(encoded_train_df,col)
to_int64(encoded_test_X,col)
```
